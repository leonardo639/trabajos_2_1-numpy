# ***# Ejercicio 03***

Crea una matriz aleatoria 7x7 y halla los valores mínimo y máximo.

```python
import numpy as np

matriz = np.random.randint(-1e3, 1e3, size=(7,7), dtype=int)

numero_minimo = np.min(matriz)
numero_maximo = np.max(matriz)

print("La matriz aleatoria 7x7 es:", end='\n\n')
print(matriz)
print("\n=================================")
print('El numero maximo de la matriz es: ',numero_maximo)
print('El numero minimo de la matriz es: ',numero_minimo)
```

Normaliza la matriz anterior entre 0 y 1.


```python
import numpy as np

matriz = np.random.randint(-1e3, 1e3, size=(7,7), dtype=int)

numero_minimo = np.min(matriz)
numero_maximo = np.max(matriz)

matriz_normalizada = (matriz + abs(numero_minimo)) / (numero_maximo + abs(numero_minimo))
np.set_printoptions(precision=2)

print("La matriz 7x7 es: \n")
print(matriz)
print("\n======================================")
print("La matriz normalizada es:", end='\n\n')
print(matriz_normalizada)
```

# ***# Ejercicio 04***

Luego invertir lo valores de 9 y 1 de la Matriz anterior.

image.png

```python
import numpy as np

matriz = np.array([[1]*5,[1,0,0,0,1],[1,0,9,0,1],[1,0,0,0,1],[1]*5])

matriz_modificada = matriz.copy()

for n,m in np.ndindex(5,5):
  if matriz_modificada[n,m] == 1:
    matriz_modificada[n,m] = 9
  elif matriz_modificada[n,m] == 9:
    matriz_modificada[n,m] = 1


print("La matriz original: ",end='n\n')
print(matriz)
print("=========================")
print("La matriz con el cambio: ", end ='\n\n')
print(matriz_modificada)
```

# ***# Ejercicio 05***

Extraer la matrices de cada color

image.png

```python
import numpy as np

matriz = np.arange(1, 31).reshape(6, 5)

#Extrayendo las matrices azul,rojo1 y rojo2; en verde solo lo convierto en una matriz 4x4 y
#haciendo una copia para no modificar a las demas matrices
matriz_azul = matriz[2:4, 0:2]
matriz_rojo1 = matriz[0:1, 3:5]
matriz_rojo2 = matriz[4:6, 3:5]
matriz_verde = matriz[0:4, 1:5].copy()

#Solo para la matriz de verde: se conserva los numeros 2,8,14 y 20 los demas se convierten en ceros.
for n,m in np.ndindex(4,4):
  if matriz_verde[n,m] == 2:
    matriz_verde[n,m] = 2
  elif matriz_verde[n,m] == 8:
    matriz_verde[n,m] = 8
  elif matriz_verde[n,m] == 14:
    matriz_verde[n,m] = 14
  elif matriz_verde[n,m] == 20:
    matriz_verde[n,m] = 20
  else:
    matriz_verde[n,m] = 0

#Impresión
print("La matriz original es: ", end='\n\n')
print(matriz)
print("\n===============================")
print("Las matrices modificadas son: ", end='\n\n')
print("MATRIZ AZUL: ")
print(matriz_azul)
print("\nMATRIZ ROJO 1: ")
print(matriz_rojo1)
print("\nMATRIZ ROJO 2: ")
print(matriz_rojo2)
print("\nMATRIZ VERDE: ")
print(matriz_verde)
```

Se debe buscar dentro de la matriz de color verde el valor = 50, luego reemplazar con la pequeña matriz de color celeste quedando la matriz de la siguiente manera:

image.png

El algoritmo solución debe poder ubicar el valor = 50 en cualquier posición y hacer el
reemplazo.

```python
#MANERA N°1
import numpy as np

matriz = np.array([[255, 34, 45, 16, 24, 99],
                   [200, 50, 67, 15, 56, 88],
                   [165, 67, 34, 26, 87, 70],
                   [120, 100, 45, 45, 56, 67],
                   [45, 225, 35, 67, 50, 44],
                   [36, 255, 24, 78, 77, 23]])

#Hacemos una copia de la matriz, llamada matriz_modificada
matriz_modificada = matriz.copy()

# Buscamos la posición del valor 50 en la matriz
n, m = np.where(matriz_modificada == 50)

# Si encontramos el valor 50, modificamos los valores alrededor a ceros
if len(n) > 0:
    for i in range(len(n)):
        matriz_modificada[n[i]-1:n[i]+2, m[i]-1:m[i]+2] = 0
        matriz_modificada[n[i],m[i]] = 255

#Imprimimos los resultados(el original como el modificado )
print('Matriz original')
print(matriz)
print("\n=============================\n")
print("Matriz con los cambios aplicados")
print(matriz_modificada)

#Problema con el codigo, si se coloca el 50 en los bordes de la matriz, no funciona como se espera
#pero si se coloca 2 50's adyacentes; uno de ellos se convertira en cero
```

```python
#MANERA N°2
import numpy as np

matriz = np.array([[255, 34, 45, 16, 24, 99],
                   [200, 50, 67, 15, 56, 88],
                   [165, 67, 34, 26, 87, 70],
                   [120, 100, 45, 45, 56, 67],
                   [45, 225, 35, 67, 50, 44],
                   [36, 255, 24, 78, 77, 23]])

# Hacemos una copia de la matriz, llamada matriz_modificada
matriz_modificada = matriz.copy()

# Buscamos la posición del valor 50 en la matriz
n, m = np.where(matriz_modificada == 50)

# Modifica los valores alrededor de 50 a cero y cambia el valor de 50 a 255
for i in range(len(n)):
    matriz_modificada[max(0, n[i]-1): min(n[i]+2, matriz.shape[0]), max(0, m[i]-1):min(m[i]+2, matriz.shape[1])] = 0
    matriz_modificada[n[i], m[i]] = 255

# Imprimimos los resultados (el original y el modificado)
print('Matriz original')
print(matriz)
print("\n=============================\n")
print("Matriz con los cambios aplicados")
print(matriz_modificada)

#Con este codigo, si se coloca el 50 en los bordes de la matriz, funciona como se espera,
#pero si se coloca 2 50's adyacentes, uno de ellos se convertira en cero
```
